```python
import pygmt
```

```python
pygmt.config(FONT_ANNOT_PRIMARY = '10p,Times-Roman',
            FONT_LABEL = '14p,Times-Roman',
            MAP_LABEL_OFFSET = '4p',
            MAP_TICK_LENGTH_PRIMARY = '5p',
            PROJ_LENGTH_UNIT = 'c',
            MAP_ANNOT_OFFSET_PRIMARY = '5p',
            MAP_ANNOT_OFFSET_SECONDARY = '5p',
            MAP_FRAME_TYPE = 'fancy',
            MAP_FRAME_WIDTH = 'auto',
            FORMAT_GEO_MAP = 'D',
            PS_CHAR_ENCODING = 'Standard+',
            MAP_FRAME_PEN = '2p,black')
```

```python
fig = pygmt.Figure()
region = [37,45.5,37,42]
fig.basemap(region=region, projection="M15c", frame = ['x1f0.5', 'y1f0.5', f'NseW'])
fig.grdimage('@earth_relief_15s',region =region ,monochrome=True, shading=True, transparency=20)
fig.coast(resolution = 'h', shorelines ='1/thinnest,black',land='white', water='grey', borders = '1/0.25p,--')
fig.show()
fig.savefig('Sandvol2003.png', dpi=300)
fig.savefig('Sandvol2003.pdf')
```
